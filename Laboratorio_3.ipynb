**Centro de Llamadas Prioritarias**

*   Se debe usar una cola para almacenar las llamadas entrantes. Cada llamada tiene un ID, un nombre de quien llama y un nivel de urgencia (de 1 a 10).

*   El programa debe realizar las siguientes operaciones:
    *   **recibirLlamada**(nombre, urgencia) → Inserta una llamada al final de la
cola.
    *   **pasarALaPila**() → Toma la llamada con mayor urgencia de la cola (si hay
varias con igual urgencia, la más antigua) y la pone en la pila de atención.
    *   **atenderLlamada**() → Atiende (elimina y muestra) la llamada que está en la
cima de la pila.
    *   **mostrarEstado**() → Muestra el estado actual de la cola y de la pila.

```python
class Cola:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.items:
            return None
        return self.items.pop(0)

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def first(self):
        if not self.items:
            return None
        return self.items[0]



class Pila:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.items:
            return None
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def top(self):
        if not self.items:
            return None
        return self.items[-1]



class Llamadas:
    def __init__(self):
        self.cola = Cola()
        self.pila = Pila()


    def recibirLlamada(self, nombre, urgencia):
        llamada = (nombre, urgencia)
        self.cola.enqueue(llamada)

    def atenderLlamada(self):
        if not self.cola.is_empty():
            llamada = self.cola.dequeue()
            self.pila.push(llamada)
            return llamada
        return None

    def pasarALaPila(self):
        if not self.cola.is_empty():
            llamada = self.cola.dequeue()
            self.pila.push(llamada)
            return llamada
        return None

    def mostrarEstado(self):
        print("Estado de la cola:")
        for item in self.cola.items:
            print(item)
        print("Estado de la pila:")
        for item in self.pila.items:
            print(item)
```

```python
llamadas = Llamadas()

llamadas.recibirLlamada("Juan", 2)
llamadas.recibirLlamada("Maria", 1)
llamadas.recibirLlamada("Pedro", 3)

llamadas.mostrarEstado()

llamadas.atenderLlamada()
llamadas.mostrarEstado()

llamadas.pasarALaPila()
llamadas.mostrarEstado()
```
